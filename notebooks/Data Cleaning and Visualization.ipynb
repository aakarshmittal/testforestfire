```python
df_copy=pd.read_csv("Algerian_cleaned.csv")
```

```python
objects=['fire','not fire']
plt.figure(figsize=(10,5))
plt.pie(df_copy['Classes'].value_counts(),labels=objects,autopct="%1.1f%%",shadow=True,explode=(0.1,0))
plt.legend(loc='upper right')
plt.show()
```
